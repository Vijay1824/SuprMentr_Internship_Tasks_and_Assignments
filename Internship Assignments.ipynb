**Create a strong code for password authentication using python.** 

```python
# Strong Password Authentication Program

def is_strong_password(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters long."

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    special_chars = "!@#$%^&*()-_=+[]{}|;:,.<>?/"

    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if not has_upper:
        return False, "Password must contain at least one uppercase letter."
    if not has_lower:
        return False, "Password must contain at least one lowercase letter."
    if not has_digit:
        return False, "Password must contain at least one digit."
    if not has_special:
        return False, "Password must contain at least one special character."

    return True, "Strong password."


def authenticate():
    correct_password = "Secure@123"   # Stored password

    while True:
        user_password = input("Enter your password: ")

        is_strong, message = is_strong_password(user_password)

        if not is_strong:
            print("❌", message)
            continue

        if user_password == correct_password:
            print("✅ Access Granted.")
            break
        else:
            print("❌ Incorrect password.")


authenticate()
```

Description : Print numbers 1–50 with Fizz/Buzz logic and count occurrences using loops and functions.

```python
def fizzbuzz():
    fizz_count = 0
    buzz_count = 0
    fizzbuzz_count = 0

    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
            fizzbuzz_count += 1
        elif i % 3 == 0:
            print("Fizz")
            fizz_count += 1
        elif i % 5 == 0:
            print("Buzz")
            buzz_count += 1
        else:
            print(i)

    print("\nCounts:")
    print("Fizz =", fizz_count)
    print("Buzz =", buzz_count)
    print("FizzBuzz =", fizzbuzz_count)

fizzbuzz()
```

# Assignment (19/02/2026)

**Assignment Name : Student Data Manager**
Description : Store data for 5 students using dictionaries, print topper, class average, and assign grades.

```python
students = {}

# Input data for 5 students
for i in range(5):
    name = input(f"Enter name of student {i+1}: ")
    marks = float(input(f"Enter marks of {name}: "))
    students[name] = marks

# Find topper
topper = max(students, key=students.get)

# Calculate class average
average = sum(students.values()) / len(students)

# Assign grades
print("\nStudent Grades:")
for name, marks in students.items():
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "Fail"

    print(name, ":", marks, "-", grade)

# Print topper and average
print("\nTopper:", topper, "-", students[topper])
print("Class Average:", average)
```

Description : Compare Python lists vs NumPy arrays with 1M numbers, measure execution time, write 3 observations.

```python
import time
import numpy as np

# Python list
python_list = list(range(1000000))

start = time.time()
python_result = [x * 2 for x in python_list]
end = time.time()

python_time = end - start

# NumPy array
numpy_array = np.arange(1000000)

start = time.time()
numpy_result = numpy_array * 2
end = time.time()

numpy_time = end - start

# Results
print("Python List Time:", python_time, "seconds")
print("NumPy Array Time:", numpy_time, "seconds")
```

# Assignment (26/02/2026)

**Assignment Name : Data Doctor**
Description : Clean a dataset by handling missing values, removing duplicates, standardizing text, and explain why cleaning matters.

```python
import pandas as pd

# Sample dataset
data = {
    "Name": ["Alice", "Bob", "alice", "Charlie", None],
    "Age": [25, 30, 25, None, 22],
    "City": ["Delhi", "Mumbai", "delhi", "Chennai", "Mumbai"]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Handle missing values
df["Name"] = df["Name"].fillna("Unknown")
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Standardize text
df["Name"] = df["Name"].str.title()
df["City"] = df["City"].str.title()

# Remove duplicates
df = df.drop_duplicates()

print("\nCleaned Dataset:")
print(df)
```

**Why Data Cleaning Matters (Explanation)**
* Missing values can cause errors or wrong analysis results.
* Duplicate rows make statistics inaccurate.
* Standardized text avoids treating similar values as different (e.g., Delhi vs delhi).
* Clean data improves machine learning accuracy.
* Good cleaning makes analysis reliable and consistent.

# Assignment (10/03/2026)

**Assignment Name : Spam Classifier Thinking**
Description : Design a spam detection system: features, data needed, possible mistakes.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample dataset
messages = [
    "Win a free prize now",
    "Hello, how are you",
    "Claim your free offer today",
    "Meeting at 5 pm",
    "Congratulations, you won money"
]

labels = ["spam", "not spam", "spam", "not spam", "spam"]

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(messages)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test new message
new_message = ["Free money offer"]
new_X = vectorizer.transform(new_message)

# Prediction
prediction = model.predict(new_X)

print("Prediction:", prediction[0])
```

**KNN in Real Life – Netflix-like Recommendation**
Explanation: 

K-nearest neighbors algorithm (K-Nearest Neighbors) works by finding items or users that are most similar.
In a recommendation system like Netflix, if two users watch similar movies, the system recommends movies liked by one user to the other.
It compares users based on:

* watched genres
* ratings
* viewing history
Then nearest similar users are selected.

Small Similarity Example
Suppose:

User A likes: Action, Sci-Fi  
User B likes: Action, Sci-Fi, Thriller  
User C likes: Romance, Drama  
 
User A is closer to User B than User C, so movies liked by User B can be recommended to User A.

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Movie preference data
# Action, Sci-Fi, Romance
data = np.array([
    [1, 1, 0],  # User A
    [1, 1, 1],  # User B
    [0, 0, 1]   # User C
])

# KNN model
model = NearestNeighbors(n_neighbors=2)
model.fit(data)

# Find nearest to User A
distances, indices = model.kneighbors([data[0]])

print("Nearest Users:", indices)
```

# Assignment (21/03/2026)

**Assignment Name : Build a Text Cleaner**
Description : Write code to remove punctuation, lowercase text, remove stopwords and test it.

```python
import string

# Simple stopwords list
stopwords = ["is", "a", "the", "and", "to", "in", "of", "it"]

# Input text
text = "This is a Simple Example, to CLEAN the text!"

# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Remove stopwords
words = text.split()
cleaned_words = [word for word in words if word not in stopwords]

# Final cleaned text
cleaned_text = " ".join(cleaned_words)

print("Original Text:", "This is a Simple Example, to CLEAN the text!")
print("Cleaned Text:", cleaned_text)
```

```python
import string
import nltk
from nltk.corpus import stopwords

# Download stopwords (run once)
nltk.download('stopwords')

# Input text
text = "This is a Simple Example, to CLEAN the text!"

# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Remove stopwords
stop_words = set(stopwords.words('english'))
words = text.split()
cleaned_words = [word for word in words if word not in stop_words]

# Final cleaned text
cleaned_text = " ".join(cleaned_words)

print("Original Text:", "This is a Simple Example, to CLEAN the text!")
print("Cleaned Text:", cleaned_text)
```
